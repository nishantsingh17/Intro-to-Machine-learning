```python
import sklearn.linear_model
import sklearn.preprocessing
import sklearn.metrics
import sklearn.model_selection
import matplotlib.pyplot as plt
```

```python
#Q.5
np.random.seed(0)
x=np.linspace(0,10,100).reshape(-1,1)
noise=np.random.normal(0,2,size=x.shape[0])
y=3*x.flatten()+5+noise
x_train, x_test, y_train, y_test = sklearn.model_selection.train_test_split(x, y, test_size=0.2, random_state=1)
lin_reg = sklearn.linear_model.LinearRegression()
lin_reg.fit(x_train, y_train)
y_pred = lin_reg.predict(x_test)
mse_1= sklearn.metrics.mean_squared_error(y_test,y_pred)
print("Mse of linear polynomial:",mse_1)
plt.scatter(x, y, color='blue', label='Data')
plt.plot(x, lin_reg.predict(x), color='red', label='Predicted')
plt.title('Linear Regression')
plt.legend()
plt.show()
```

```python
#Q.5
np.random.seed(0)
x=np.linspace(0,1,100).reshape(-1,1)
noise=np.random.normal(0,2,size=x.shape[0])
y=3*x.flatten()**2+2*x.flatten()+5+noise
x_train, x_test, y_train, y_test = sklearn.model_selection.train_test_split(x, y, test_size=0.2, random_state=1)
lin_reg = sklearn.linear_model.LinearRegression()
lin_reg.fit(x_train, y_train)
y_pred = lin_reg.predict(x_test)
mse= sklearn.metrics.mean_squared_error(y_test,y_pred)
print("Mse of quadratic polynomial:",mse)
plt.scatter(x, y, color='blue', label='Data')
plt.plot(x, lin_reg.predict(x), color='red', label='Predicted')
plt.title('Linear Regression')
plt.legend()
plt.show()
```

```python
#Q.5
np.random.seed(0)
x=np.linspace(0,1,100).reshape(-1,1)
noise=np.random.normal(0,2,size=x.shape[0])
y=4*x.flatten()**3+5*x.flatten()**2+9*x.flatten()+5+noise
x_train, x_test, y_train, y_test = sklearn.model_selection.train_test_split(x, y, test_size=0.2, random_state=1)
lin_reg = sklearn.linear_model.LinearRegression()
lin_reg.fit(x_train, y_train)
y_pred = lin_reg.predict(x_test)
mse= sklearn.metrics.mean_squared_error(y_test,y_pred)
print("Mse of cubic polynomial:",mse)
plt.scatter(x, y, color='blue', label='Data')
plt.plot(x, lin_reg.predict(x), color='red', label='Predicted')
plt.title('Linear Regression')
plt.legend()
plt.show()
```
